```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from itertools import combinations
from scipy.stats import ttest_ind

import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_rows', 50) # показывать больше строк
pd.set_option('display.max_columns', 50) # показывать больше колонок

stud = pd.read_csv('stud_math.csv')
display(stud.head(10))
stud.info()

#В датасете 13 числовых столбцов
#В датасете 17 строковых столбцов
```

```python
#Описание функций, которые будут применяться к колонкам исследуемого дата-сета

#Функция, которая возвращает данные колонки, по переданным в качестве переменных имени дата-сета и имени колонки
def value_count(file_name, col_name):
    print('Данные колонки: ', col_name, '/////////')
    result = pd.DataFrame(file_name[col_name].value_counts())
    return result

#Функция, которая выводит статистические данные колонки, по переданным в качестве переменных имени дата-сета и имени колонки
def stat_count(file_name, col_name):
    print('Статистические данные колонки: ', col_name,'///////////')
    print(file_name[col_name].describe())

#Функция, которая выводит данные колонки, по переданным в качестве переменных имени дата-сета и имени колонки
#в графическом виде, в виде столбчатой диаграммы
def stat_graf(file_name, col_name):
    print('График : ', col_name,'///////////')
    print(file_name[col_name].plot(kind = 'hist', 
                               grid = True, 
                               subplots = True, 
                               title = [col_name], 
                               legend = True)) 
    plt.show()
    
#Функция, которая выводит данные о существующих выбросах колонки, 
#по переданным в качестве переменных имени дата-сета и имени колонки
#и удаляет их    
def stat_iqr(file_name, col_name):
    print('Данные о выбросах колонки: ', col_name,'///////////')
    IQR = file_name[col_name].quantile(0.75) - file_name[col_name].quantile(0.25)
    perc25 = file_name[col_name].quantile(0.25)
    perc75 = file_name[col_name].quantile(0.75)
 
    print(col_name,
    '25-й перцентиль: {},'.format(perc25),
    '75-й перцентиль: {},'.format(perc75),
    "IQR: {}, ".format(IQR),
    "Границы выбросов: [{f}, {l}].".format(f=perc25 - 1.5*IQR, l=perc75 + 1.5*IQR))
    
    file_name = file_name.loc[file_name[col_name].between(perc25 - 1.5*IQR, perc75 + 1.5*IQR)]

#Функция, которая очищает данные колонки(содержащих числовые значения) от пустых значений, 
#по переданным в качестве переменных имени дата-сета и имени колонки    
def del_nan(column):
    column = column.fillna(0)
    print('Очищенные данные колонки: ', column,'///////////')
    print(column.unique())
    
#Функция, которая очищает данные колонки(содержащих нечисловые данные) от пустых значений, 
#по переданным в качестве переменных имени дата-сета и имени колонки 
#и возвращает количество уникальных значений для данных этой колонки
def del_nan_nom(column):
    column = column.astype(str).apply(lambda x: 'no_answer' if x.strip() == 'nan' else x)
    print('Очищенные данные колонки: ', column,'///////////')
    print("Уникальных значений:", column.nunique())
    
    
#Функция, которая выводит данные колонки(содержащих номинативные значение) в графическом виде
#(в виде диаграммы размаха или ящика с усами), 
#по переданным в качестве переменных имени дата-сета и имени колонки    
def get_boxplot(column):
    print('Графические данные колонки: ', column,'///////////')
    fig, ax = plt.subplots(figsize = (14, 4))
    sns.boxplot(x=column, y='score', 
                data=stud.loc[stud.loc[:, column].isin(stud.loc[:, column].value_counts().index[:10])],
               ax=ax)
    plt.xticks(rotation=45)
    ax.set_title('Boxplot for ' + column)
    plt.show()  
    
#Функция, которая выводит статистически значимые различия в данных колонки(содержащих номинативные значения), 
#по переданным в качестве переменных имени дата-сета и имени колонки    
def get_stat_dif(column):
    print('Статистически значимые различия колонки: ', column,'///////////')
    cols = stud.loc[:, column].value_counts().index[:10]
    combinations_all = list(combinations(cols, 2))
    for comb in combinations_all:
        if ttest_ind(stud.loc[stud.loc[:, column] == comb[0], 'score'], 
                        stud.loc[stud.loc[:, column] == comb[1], 'score']).pvalue \
            <= 0.05/len(combinations_all): # Учли поправку Бонферони
            print('Найдены статистически значимые различия для колонки', column)
            break    
    
pd.set_option('display.max_rows', 50) # показывать больше строк
pd.set_option('display.max_columns', 50) # показывать больше колонок

stud = pd.read_csv('stud_math.csv')
display(stud.head(10))
stud.info()

#Получаем список всех колонок дата-сета
columns = list(stud)

#В цикле обходим колонки дата-сета и по каждой выводим ее данные и сводную информацию по этим данным
for column in columns:
    print(value_count(stud, column))
    print(stud[column].unique())

#Получаем словарь из всех числовых колонок дата-сета, где пара 'ключ-значение' - это 'псевдоним - имя колонки'   
int_col = {'Возраст ученика':'age', 'Образование матери':'Medu', 'Образование отца':'Fedu', 'Время в пути до школы':'traveltime', 'Время на учебу вне школы':'studytime','Количество внеучебных неудач':'failures', 'Семейные отношения':'famrel', 'Свободное время':'freetime', 'Время с друзьями':'goout', 'Здоровье':'health','Кол-во пропусков занятий':'absences', 'Баллы за ГЭ по математике':'score'}

#В цикле обходим числовые колонки и обрабаотываем их данные
for i_col in int_col:
    
    #Исследуем статистические данные
    stat_count(stud, int_col[i_col])
    #Очищаем от пустых значений
    del_nan(stud[int_col[i_col]])
    #Выводим график распределения данных
    stat_graf(stud, int_col[i_col])
    #исследуем на наличие выбросов в данных и удалим их
    stat_iqr(stud, int_col[i_col])
    
#Построим графики зависимости переменных
print(sns.pairplot(stud, kind = 'reg'))

#Производим корреляционный анализ, выводим матрицу корреляций
print(stud.corr())

#Данные колонки "studytime, granular" находятся в полной обратной корреляции к данным колонки "studytime",
#являются избыточными, потому мб удалены 
stud.drop(['studytime, granular'], axis = 'columns', inplace = True)

#Получаем словарь из всех колонок дата-сета, содержащих номинативные данные, где пара 'ключ-значение' - это 'псевдоним - имя колонки'
obj_col = {'Школа':'school', 'Размер семьи':'famsize','Статус родителей':'Pstatus', 'Работа матери':'Mjob', 'Работа отца':'Fjob','Пол':'sex', 'Адрес':'address', 'Причина выбора школы':'reason','Опекун':'guardian', 'Доп.образование':'schoolsup','Семейная поддержка':'famsup', 'Доп.занятия по математике':'paid','Доп.занятия':'activities', 'Посещение ДОУ':'nursery','Планы ВО':'higher', 'Домашний интернет':'internet','Романтические отношения':'romantic'}

#В цикле обходим нечисловые колонки и обрабатываем их данные

for ob_col in obj_col:
    
    #Очищаем от пустых значений и находим уникальные значения
    del_nan_nom(stud[obj_col[ob_col]])
    #Рассмотрим плотность распределения номинативных данных
    get_boxplot(obj_col[ob_col])
    #Найдем возможные статистически значимые различия в данных по способу Стьюдента
    get_stat_dif(obj_col[ob_col])
```

Оценка качества данных дата-сета:

 - выборка не очень большая, а потому данный анализ нельзя в полной мере назвать объективным, но для учебной задачи данных достаточно
 - показателей достаточно много для целостного отражения влияния их данных на предсказываемую величину, для их правильной идентификации и взаимосвязи.
 - у всех показателей диапазон принимаемых значений ограничен и не имеет существенных выбросов, что позволяет в должной мере оценивать их влияние на предсказываемую величину.
 - данные дата-сета не слишком коррелируют друг с другом, а значит не являются избыточными и дб использованы при анализе;
 - удаление всех строк, имеющих пустые значения, не представляется возможным, чтобы сохранить релевантность данных.

Таким образом, дата-сет может быть использован для учебных задач, а используемый для анализа функционал мб использован для исследования большего обьема данных.

В нашем случае мы исследуем влияние различных показателей на оценку по математике среднего ученика школы (то есть зависимая величина это показатель score)

На основании произведенной обработки данных дата-сета можно сделать следующие выводы:

Основная масса учеников:

 - находится в возрасте от 15 до 19 лет
 - проживает в городе
 - растет в многодетных семьях
 - родители проживают вместе
 - матери в семьях имеют более высокое образование чем отцы
 - родители чаще заняты в службах сервиса либо в предприятиях не относящихся к социальным службам
 - основным опекуном является мать
 - живут вблизи школы
 - учатся допольнительно
 - успешны во внеучебное время
 - посещали ДОУ
 - планируют получать ВО
 - имеют хорошие семейные отношения
 - имеют достаточно свободного времени
 - имеют достаточно времени на общение с друзьями
 - имеют хорошее здоровье
 - мало пропускают занятия


Более высокое влияние на оценку имеет:

 - возраст ученика (чем старше дети, тем хуже учатся)
 - образование матери (чем выше образование матери, тем лучше учатся дети)
 - количество внеучебных неудач (чем более успешен ребенок вне школы (меньше количество неудач), тем лучше учится)
 - количество проводимого времени с друзьями (чем больше веселится ребенок с друзьями, тем хуже учится)

 - пол ученика (мальчики учатся лучше (хм..))
 - место проживания (городские дети учатся лучше)
 - работа родителей (дети учителей учатся лучше)
 - статус опекуна (с родителями дети учатся лучше)
 - дополнительная образовательная поддержка (логично, более сильные ученики в доп.помощи нуждаются реже)
 - дополнительное образование (лучше учатся те, кто не занимаются дополнительно)
 - планы на получение ВО (учатся лучше те, кто планируют продолжать образование в ВУЗах)
 - романтические отношения (лучше учатся дети, свободные от отношений)

Практически не влияет на оценку:

 - тип школы
 - причина выбора школы
 - размер семьи
 - статус совместного проживания родителей
 - поддержка в семье
 - посещение ДОУ
 - наличие домашнего интернета

Если посмотреть на матрицу корреляций, то нет сильно скоррелированных данных, за исключением колонок "studytime, granular" и "studytime", одну из которых можно удалить. Остальные данные необходимо исследовать для получения корректных данных.
